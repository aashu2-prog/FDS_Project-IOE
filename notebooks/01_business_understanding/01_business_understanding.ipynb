# Stage 1 — Business Understanding
**Project:** Driver Warning System — Nepal Road Accident Analysis  
**Date:** 2026-03-03

## 1.1 Problem Statement

Road accidents in Nepal cause significant loss of life and property every year. Drivers often lack real-time awareness of how dangerous their current location or driving conditions are.

**Goal:** Build a system that, given a driver's location (Province/District) and contextual conditions (time of day, weather, road type), predicts the **Accident Severity Zone** and issues an appropriate warning.

## 1.2 Business Objectives

| # | Objective |
|---|---|
| 1 | Identify high-risk provinces and districts |
| 2 | Understand which conditions (weather, road, time) increase severity |
| 3 | Build a classifier that predicts severity level for a given region + conditions |
| 4 | Use predictions to warn drivers before entering high-risk zones |

## 1.3 Success Criteria

- Model accuracy ≥ 60% on multi-class severity prediction  
- Clear visualisations showing which regions/conditions are most dangerous  
- A working prediction function: input location + conditions → severity warning

## 1.4 Data Science Lifecycle Plan

```
Stage 1 → Business Understanding   (this notebook)
Stage 2 → Data Acquisition         (load & inspect raw data)
Stage 3 → Data Cleaning            (fix missing, duplicates, typos)
Stage 4 → EDA                      (visualise patterns & relationships)
Stage 5 → Feature Engineering      (encode & prepare for modelling)
Stage 6 → Modelling                (train classifier)
Stage 7 → Evaluation               (metrics, confusion matrix)
```
