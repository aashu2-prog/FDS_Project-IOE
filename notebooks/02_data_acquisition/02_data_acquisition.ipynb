# Stage 2 — Data Acquisition
**Goal:** Load the raw dataset and understand its structure before any cleaning.

```python
import pandas as pd
import numpy as np
from pathlib import Path

RAW_DIR = Path('../../data/01_raw')

df = pd.read_csv(RAW_DIR / 'dataset.csv')
print('Shape:', df.shape)
df.head(10)
```

## 2.1 Column Descriptions

| Column | Type | Description |
|---|---|---|
| Accident_ID | int | Unique identifier |
| Accident_Date | str | Date of accident |
| Year | int | Year extracted from date |
| Month | str | Month name |
| Province | str | One of 7 provinces of Nepal |
| District | str | District name |
| Road_Type | str | Highway / Urban / Rural / Feeder |
| Intersection_Type | str | Straight / Curve / Junction |
| Time_of_Day | str | Morning / Afternoon / Evening / Night |
| Accident_Severity | str | **Target** — Fatal / Serious Injury / Minor Injury / Damage Only |
| Weather_Condition | str | Clear / Rain / Fog |
| Road_Surface | str | Dry / Wet / Slippery / Damaged |
| Vehicle_Type | str | Car / Motorbike / Truck / Bus / Scooter / Microbus |
| Speed_Zone | int | Speed limit in km/h |
| Driver_Gender | str | Male / Female |

```python
# Basic data types and non-null counts
df.info()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Missing value count per column
missing = df.isnull().sum()
print('Missing values:\n', missing[missing > 0])
```

```python
# Duplicate rows
print('Duplicate rows:', df.duplicated().sum())
```

```python
# Unique values in target column — check for label inconsistencies
print('Accident_Severity unique values:')
print(df['Accident_Severity'].value_counts())
```

```python
# Check Speed_Zone for outliers
print('Speed_Zone unique values:', sorted(df['Speed_Zone'].dropna().unique()))
```

```python
# Check Province for casing inconsistencies
print('Province unique values:', sorted(df['Province'].dropna().unique()))
```

```python
# Check Accident_Date for corrupted entries
bad_dates = df[~df['Accident_Date'].astype(str).str.match(r'\d{4}-\d{2}-\d{2}')]
print(f'Corrupted date rows: {len(bad_dates)}')
print(bad_dates[['Accident_ID','Accident_Date']].head())
```

## 2.2 Data Quality Issues Found

| Issue | Columns Affected |
|---|---|
| Missing values | Weather_Condition, Road_Surface, Driver_Gender, District, Time_of_Day |
| Duplicate rows | ~32 rows |
| Typos in labels | Accident_Severity (e.g. 'Fatall', 'minor injury') |
| Inconsistent casing | Province (some ALL CAPS) |
| Outlier values | Speed_Zone = 999 |
| Year mismatch | Year = 1999 but Accident_Date is recent |
| Corrupted dates | Accident_Date = 'N/A' |

➡ All issues will be fixed in **Stage 3 — Data Cleaning**.
